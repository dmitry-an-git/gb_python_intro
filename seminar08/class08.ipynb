```python
#  Задача No49. Решение в группах
# Создать телефонный справочник с возможностью импорта и экспорта данных в формате .txt.
# Фамилия, имя, отчество, номер телефона - данные, которые должны находиться в файле.
# 1. Программа должна выводить данные
# 2. Программа должна сохранять данные в
# текстовом файле
# 3. Пользователь может ввести одну из
# характеристик для поиска определенной записи(Например имя или фамилию человека)
# 4. Использование функций. Ваша программа не должна быть линейной

# a - открытие для добавления (и создание если пустой)
# r - открытие для чтения
# w - открытие для перезаписи (и создание)
# w+ - открытие для чтения и перезаписи (и создание)
# r+ - чтение и добавление
```

```python
lines = ["some new line here","second new string here"]
with open("data.txt", "w", encoding = "UTF-8") as data:
    for line in lines:
        data.write(line)
        data.write("\n")

with open("data.txt", "r") as data:
    lines = data.readlines()
print(lines)
```

```python
def tup_to_txt(tup):
    txt = ""
    for i in tup:
        txt += i
        txt += ";"
    txt += "\n"
    return txt

def txt_to_tup(txt):
    return tuple(txt.split(";")[:-1])
    
def load_lines():
    print("Reading the DB...", end = " ")
    lines = []
    with open("data.txt", "r") as data:
        lines = data.readlines()
        lines = [txt_to_tup(line) for line in lines]
    print("Done")
    if len(lines)==0:
        print("Warning! The DB is empty!")
    return lines
    
def enter_new():
    name = input("Please enter the first name: ")
    surname = input("Please enter the second name: ")
    phone = input("Please enter the phone number: ")
    return (name,surname,phone)

def add_new(entry):
    lines.append(entry)
        
def print_line(line):
    for item in line:
        print(item, end = "\t\t")
    print("")

def print_all(db):
    print()
    if len(db) == 0:
        print("The DB is empty! Nothing to print!")
    else: 
        print("start of DB")
        for index, line in enumerate(db):
            print(index, end = "\t\t")
            print_line(line)
        print("end of DB")
            
def new_db():
    print()
    with open("data.txt", "w") as data:
        data.write("")
    print("The db was cleared!")
    load_lines()
        
def search_txt():
    prompt = input("Please enter the keyword to search: ")
    result = []
    for i in range(len(lines)):
        for j in range(len(lines[i])):
            if lines[i][j] == prompt:
                result.append(i)
    result = set(result)
    if len(result) == 0:
        print("Nothing is found")
    else: 
        print("The following is found:")
        for i in result:
            print(i, end = "\t\t")
            print_line(lines[i])
            
def modify():
    size = 0
    with open("data.txt", "r") as data:
        lines = data.readlines()
        size = len(lines)
    index = int(input("Please enter the index of the line you want to modify: "))
    if index>size-1:
        return print("Wrong index!")
    else: 
        txt = ""
        new_entry = enter_new()
        for i in new_entry:
            txt += i
            txt += ";"
        txt += "\n"
        lines[index] = txt
        
    with open("data.txt", "w") as data:
        data.write(str(lines))
        
def save():
    with open("data.txt", "w") as data:
        for i in lines:
            data.write(tup_to_txt(i))
    print("The changes are saved successfully.")
        
def main():
    lines = load_lines()
    print('The following commands are accepted: \n"a" - add one entry \n"p" - print all entries \n"f" - search \n"m" - modify one entry \n"d" - erase all \n"s" - save changes to disk \n"q" - exit')
    while True:
        print("----")
        command = input("Please enter the command: ")
        if command == "a":
            add_new(enter_new())
        elif command == "p":
            print_all(lines)
        elif command == "f":
            search_txt()
        elif command == "m":
            modify()
        elif command == "d":
            new_db()
        elif command == "s":
            save()
        elif command == "q":
            break
        else: 
            print("Wrong command!")
            continue
    
    
lines = []
main()
    
```
